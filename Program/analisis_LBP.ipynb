```python
import os
import pandas as pd

# Folder where the CSV files are located
folder_path = r'D:\Zidan\Citra Digital\Pengenalan pola\outputs\LBP'

# List to hold the DataFrames
dataframes = []

# Get a list of all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Ensure we only process up to 8 files
for i, filename in enumerate(csv_files[:8]):
    file_path = os.path.join(folder_path, filename)
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    dataframes.append(df)

# Optionally: Create separate variables for each DataFrame if needed
df1, df2= dataframes[:2]
```

```python
def process_dataframe(df):
    """Drop non-numeric columns and compute the mean of the remaining columns."""
    df_numeric = df.drop(columns=['filename'])  # Dropping the 'filename' column
    return df_numeric.mean()  # Returning the mean of each column

# Process all the sample dataframes
mean_tapioka = process_dataframe(df1)
mean_terigu = process_dataframe(df2)

# Combine all the means into a single dataframe
mean_df = pd.DataFrame({
    'tapioka': mean_tapioka,
    'terigu': mean_terigu,
})

mean_df
```

```python
import matplotlib.pyplot as plt

# Daftar fitur LBP yang ingin dibandingkan (ganti sesuai kolom mean_df Anda)
variables = ['contrast_R', 'contrast_G', 'contrast_B',
             'dissimilarity_R', 'dissimilarity_G', 'dissimilarity_B',
             'homogeneity_R', 'homogeneity_G', 'homogeneity_B',
             'energy_R', 'energy_G', 'energy_B',
             'correlation_R', 'correlation_G', 'correlation_B',
             'entropy_R', 'entropy_G', 'entropy_B']

# Untuk setiap fitur utama (tanpa channel)
fitur_utama = ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation', 'entropy']
channel_labels = ['Red', 'Green', 'Blue']
channel_codes = ['R', 'G', 'B']

for fitur in fitur_utama:
    plt.figure(figsize=(8, 5))
    # Ambil mean untuk tiap channel
    tapioka_values = [mean_df.loc[f'{fitur}_{ch}', 'tapioka'] for ch in channel_codes]
    terigu_values = [mean_df.loc[f'{fitur}_{ch}', 'terigu'] for ch in channel_codes]
    
    # Plot
    plt.plot(channel_labels, tapioka_values, marker='o', linestyle='-', color='#66c2ff', linewidth=2, label='Tapioka')
    plt.plot(channel_labels, terigu_values, marker='o', linestyle='-', color='#ff6666', linewidth=2, label='Terigu')
    
    # Tampilkan nilai di tiap titik
    for j, val in enumerate(tapioka_values):
        plt.annotate(f'{val:.4f}', (channel_labels[j], val), textcoords="offset points", xytext=(0,5), ha='center', fontsize=8, color='#0066cc')
    for j, val in enumerate(terigu_values):
        plt.annotate(f'{val:.4f}', (channel_labels[j], val), textcoords="offset points", xytext=(0,-15), ha='center', fontsize=8, color='#cc0000')
    
    plt.grid(True, linestyle='-', alpha=0.7)
    plt.title(f'Rata-rata {fitur.capitalize()} LBP', fontsize=12)
    plt.xlabel('Kanal Warna (RGB)', fontsize=10)
    plt.ylabel(f'Mean {fitur.capitalize()}', fontsize=10)
    plt.legend(loc='upper left', fontsize=9)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Daftar fitur utama dan channel
fitur_utama = ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation', 'entropy']
channel_labels = ['Red', 'Green', 'Blue']
channel_codes = ['R', 'G', 'B']

# Hitung standar deviasi untuk tiap fitur dan channel
std_tapioka = df1.drop(columns=['filename']).std()
std_terigu = df2.drop(columns=['filename']).std()

n_fitur = len(fitur_utama)
n_cols = 2
n_rows = (n_fitur + 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 3 * n_rows))
axes = axes.flatten()

for i, fitur in enumerate(fitur_utama):
    ax = axes[i]
    # Ambil std deviasi untuk tiap channel
    tapioka_values = [std_tapioka[f'{fitur}_{ch}'] for ch in channel_codes]
    terigu_values = [std_terigu[f'{fitur}_{ch}'] for ch in channel_codes]
    
    # Plot
    ax.plot(channel_labels, tapioka_values, marker='o', linestyle='-', color='#66c2ff', linewidth=2, label='Tapioka')
    ax.plot(channel_labels, terigu_values, marker='o', linestyle='-', color='#ff6666', linewidth=2, label='Terigu')
    
    # Tampilkan nilai di tiap titik
    for j, val in enumerate(tapioka_values):
        ax.annotate(f'{val:.4f}', (channel_labels[j], val), textcoords="offset points", xytext=(0,5), ha='center', fontsize=8, color='#0066cc')
    for j, val in enumerate(terigu_values):
        ax.annotate(f'{val:.4f}', (channel_labels[j], val), textcoords="offset points", xytext=(0,-15), ha='center', fontsize=8, color='#cc0000')
    
    ax.grid(True, linestyle='-', alpha=0.7)
    ax.set_title(f'Std Deviasi {fitur.capitalize()} LBP', fontsize=11)
    ax.set_xlabel('Kanal Warna (RGB)', fontsize=9)
    ax.set_ylabel(f'Std Deviasi {fitur.capitalize()}', fontsize=9)
    ax.legend(loc='upper left', fontsize=8)

# Hilangkan subplot kosong jika ada
for i in range(n_fitur, len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

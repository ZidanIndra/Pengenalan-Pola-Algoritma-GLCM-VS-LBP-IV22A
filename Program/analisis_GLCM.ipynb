```python
import os
import pandas as pd

# Folder where the CSV files are located
folder_path = r'D:\Zidan\Citra Digital\Pengenalan pola\outputs'

# List to hold the DataFrames
dataframes = []

# Get a list of all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Ensure we only process up to 8 files
for i, filename in enumerate(csv_files[:8]):
    file_path = os.path.join(folder_path, filename)
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    dataframes.append(df)

# Optionally: Create separate variables for each DataFrame if needed
df1, df2= dataframes[:2]
```

```python
def process_dataframe(df):
    """Drop non-numeric columns and compute the mean of the remaining columns."""
    df_numeric = df.drop(columns=['filename'])  # Dropping the 'filename' column
    return df_numeric.mean()  # Returning the mean of each column

# Process all the sample dataframes
mean_sesi1 = process_dataframe(df1)
mean_sesi2 = process_dataframe(df2)

# Combine all the means into a single dataframe
mean_df = pd.DataFrame({
    'sesi1': mean_sesi1,
    'sesi2': mean_sesi2,
})

mean_df
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the top-level categories (grouped labels) and their respective subcategories
concentrations = ['Baru', '4 jam', '8 jam', '12 jam', '16 jam','20 jam','24 jam','28 jam']
channels = ['Blue', 'Green', 'Red']

# The number of channels per concentration level
n_channels = len(channels)

# Create x-axis labels and positions for the 6 values (3 channels × 2 sessions)
x_labels = ['Blue sesi1', 'Green sesi1', 'Red sesi1', 'Blue sesi2', 'Green sesi2', 'Red sesi2']
tick_positions = np.arange(len(x_labels))

# Extract the feature data for each color channel and percentage
contrast_blue = mean_df.loc['contrast_B']
contrast_green = mean_df.loc['contrast_G']
contrast_red = mean_df.loc['contrast_R']

homogeneity_blue = mean_df.loc['homogeneity_B']
homogeneity_green = mean_df.loc['homogeneity_G']
homogeneity_red = mean_df.loc['homogeneity_R']

# Combine the channels for Contrast and Homogeneity
contrast = [contrast_blue[0], contrast_green[0], contrast_red[0],
            contrast_blue[1], contrast_green[1], contrast_red[1],]

homogeneity = [homogeneity_blue[0], homogeneity_green[0], homogeneity_red[0],
               homogeneity_blue[1], homogeneity_green[1], homogeneity_red[1],]

# Plotting Contrast
plt.figure(figsize=(25, 10))

plt.subplot(2, 1, 1)
plt.plot(tick_positions, contrast, label='Contrast', marker='o', color='blue')
plt.title('Contrast Across Blue, Green, Red Channels')
plt.ylabel('Average Values')
plt.grid(True)

# Set the bottom-level ticks for channels
plt.xticks(tick_positions, x_labels, rotation=0)

# Add the top-level concentration labels
for i, concentration in enumerate(concentrations):
    plt.text(i * n_channels + 1, plt.ylim()[0] - 0.0001, concentration, ha='center', va='top', fontsize=10, fontweight='bold')

# Add vertical separator lines between concentration levels
for i in range(1, len(concentrations)):
    plt.axvline(x=i * n_channels - 0.5, color='black', linestyle='--', lw=1)

# Plotting Homogeneity (Second Subplot)
plt.subplot(2, 1, 2)
plt.plot(tick_positions, homogeneity, label='Homogeneity', marker='o', color='orange')
plt.title('Homogeneity Across Blue, Green, Red Channels')
plt.ylabel('Average Values')
plt.grid(True)

# Set the bottom-level ticks for channels
plt.xticks(tick_positions, x_labels, rotation=0)

# Add the top-level concentration labels for the second subplot
for i, concentration in enumerate(concentrations):
    plt.text(i * n_channels + 1, plt.ylim()[0] - 0.00005, concentration, ha='center', va='top', fontsize=10, fontweight='bold')

# Add vertical separator lines between concentration levels
for i in range(1, len(concentrations)):
    plt.axvline(x=i * n_channels - 0.5, color='black', linestyle='--', lw=1)

plt.tight_layout()
plt.show()

plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Load data dari file CSV
df_tapioka = pd.read_csv(r'D:\Zidan\Citra Digital\Pengenalan pola\outputs\Tapioka_glcm_result.csv')
df_terigu = pd.read_csv(r'D:\Zidan\Citra Digital\Pengenalan pola\outputs\Terigu_glcm_result.csv')

# Identifikasi kolom GLCM (yang mengandung '_')
glcm_columns = [col for col in df_tapioka.columns if '_' in col]

# Hitung standar deviasi dari tiap kolom
std_tapioka = df_tapioka[glcm_columns].std()
std_terigu = df_terigu[glcm_columns].std()

# Ambil channel dan fitur dari nama kolom
features = set(col.split('_')[1] for col in glcm_columns)
channels = set(col.split('_')[0] for col in glcm_columns)

# Buat folder output
output_folder = 'gambar/grafik std'
os.makedirs(output_folder, exist_ok=True)

# Buat grafik batang standar deviasi
for channel in channels:
    for feature in features:
        col = f'{channel}_{feature}'
        if col in std_tapioka and col in std_terigu:
            plt.figure()
            plt.bar(['Tapioka', 'Terigu'], [std_tapioka[col], std_terigu[col]],
                    color=['lightblue', 'lightsalmon'])
            plt.title(f'Standar Deviasi GLCM - {channel} {feature}')
            plt.ylabel('Standar Deviasi')
            plt.tight_layout()
            
            # Simpan grafik
            filename = f'{channel}_{feature}_std.png'
            plt.savefig(os.path.join(output_folder, filename))
            plt.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

# Membaca file CSV
tapioka_data = pd.read_csv(r'D:\Zidan\Citra Digital\Pengenalan pola\outputs\Tapioka_glcm_result.csv')
terigu_data = pd.read_csv(r'D:\Zidan\Citra Digital\Pengenalan pola\outputs\Terigu_glcm_result.csv')

# Daftar variabel yang akan dianalisis
variables = ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation', 'asm', 'entropy']

# Daftar channel dengan label yang lebih deskriptif
channel_labels = ['Red', 'Green', 'Blue']
channel_codes = ['R', 'G', 'B']

# Menghitung ukuran subplot layout yang sesuai
n_vars = len(variables)
n_cols = 2  # Jumlah kolom dalam subplot
n_rows = math.ceil(n_vars / n_cols)  # Menghitung jumlah baris yang diperlukan

# Membuat figure dan subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 4*n_rows))
axes = axes.flatten()  # Mengubah array 2D menjadi 1D untuk akses yang lebih mudah

# Membuat plot untuk setiap variabel
for i, variable in enumerate(variables):
    ax = axes[i]
    
    # Data untuk setiap channel (menggunakan std deviation)
    tapioka_values = [tapioka_data[f'{variable}_{ch}'].std() for ch in channel_codes]
    terigu_values = [terigu_data[f'{variable}_{ch}'].std() for ch in channel_codes]
    
    # Plot untuk setiap jenis tepung dengan style sesuai contoh
    tapioka_line = ax.plot(channel_labels, tapioka_values, marker='o', linestyle='-', 
                          color='#66c2ff', linewidth=2, label='Tapioka')
    terigu_line = ax.plot(channel_labels, terigu_values, marker='o', linestyle='-', 
                         color='#ff6666', linewidth=2, label='Terigu')
    
    # Menampilkan nilai pada setiap titik
    for j, val in enumerate(tapioka_values):
        ax.annotate(f'{val:.4f}', (channel_labels[j], val), 
                   textcoords="offset points", xytext=(0,5), ha='center',
                   fontsize=8, color='#0066cc')
    
    for j, val in enumerate(terigu_values):
        ax.annotate(f'{val:.4f}', (channel_labels[j], val), 
                   textcoords="offset points", xytext=(0,-15), ha='center',
                   fontsize=8, color='#cc0000')
    
    # Mengatur grid seperti contoh
    ax.grid(True, linestyle='-', alpha=0.7)
    
    # Mengatur judul dan label
    ax.set_title(f'Nilai Standar Deviasi {variable.capitalize()} GLCM', fontsize=11)
    ax.set_xlabel('Kanal Warna (RGB)', fontsize=9)
    ax.set_ylabel(f'Standar Deviasi {variable.capitalize()}', fontsize=9)
    
    # Menambahkan legend di pojok kiri atas
    ax.legend(loc='upper left', fontsize=8)

# Menghilangkan subplot kosong jika ada
for i in range(n_vars, len(axes)):
    fig.delaxes(axes[i])

# Menyesuaikan tampilan
plt.tight_layout()
plt.subplots_adjust(hspace=0.4, wspace=0.3)  # Menambah ruang antar subplot

# Menampilkan semua plot
plt.show()

# Jika ingin disimpan
# plt.savefig('glcm_features_comparison.png', dpi=300, bbox_inches='tight')
```
